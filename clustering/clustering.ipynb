```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
stocks=pd.read_csv('data_stocks.csv',index_col=0)
```

```python
stock_test=stocks.copy()
```

```python
stock_test.index=pd.to_datetime(stock_test.index)
```

```python
stock_test=stock_test.resample('100U').mean()
stock_test.fillna(method='ffill',inplace=True)
```

```python
scaled=minmax_scale(stock_test)
x_s=pd.DataFrame(scaled)
```

```python
x_trans=x_s.transpose()
x_trans
```

```python
plt.plot(stock_test.iloc[:,0])
```

```python
plt.plot(stocks.iloc[:,0])
```

```python
plt.plot(x_trans.iloc[0,:])
```

```python
from sklearn.preprocessing import minmax_scale
from sklearn.model_selection import train_test_split
from sklearn.cluster import *
from sklearn.metrics import *
```

```python
x_train,x_test=train_test_split(x_trans,test_size=0.10)
```

```python
x_train
```

```python
x_test.head()
```

```python
x_train.describe()
```

```python
wcss=[]
for i in range(1,80):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=9)
    kmeans.fit(x_train)
    wcss.append(kmeans.inertia_)
    
```

```python
plt.plot(range(1,80),wcss,'r+')
plt.plot(range(1,80),wcss)
plt.grid()
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
k_cluster=KMeans(n_clusters=30,init='k-means++')
k_cluster.fit(x_train)
```

```python
labels_k=k_cluster.labels_
```

```python
np.where(labels_k==18)
```

```python
labels_k
```

```python
silhouette_score(x_train,k_cluster.labels_)
```

```python
plt.figure(figsize=(10,10))
pallette=sns.color_palette("Spectral", as_cmap=True)
sns.scatterplot(x=x_train.columns[0],y=x_train.columns[-1],data=x_train,hue=labels_k,palette=pallette)
plt.plot([0,1],[0,1],'r')
```

```python
x_te_cl=k_cluster.predict(x_test)
x_te_cl
```

```python
np.where(x_te_cl==1)
```
